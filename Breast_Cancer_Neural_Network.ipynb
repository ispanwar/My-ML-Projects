```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
# loading the data to a data frame
```

```python
data.head()
```

```python
data['label'] = breast_cancer_dataset.target
```

```python
X = data.drop(columns='label',axis = 1)
y = data['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation= 'relu'),
    keras.layers.Dense(2,activation = 'sigmoid')]
)
```

```python
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,
                    validation_split = 0.1,
                    epochs = 10)
```

```python
history = model.fit(X_train_std,y_train,validation_split=0.1,epochs = 10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training data','validataion data'],loc = 'lower right')
```

```python
loss, accuracy = model.evaluate(X_test_std,y_test)
```

```python
print(accuracy*100)
```
